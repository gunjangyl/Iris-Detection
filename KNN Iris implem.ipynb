```python
#Separating dataset into independnt & dependent variables
#Extracting independent variables(sepal-length, sepal-width, petal-lemgth, petal-width)
X=df.iloc[:,:4].values
X
```

```python
#Extracting dependent variable y(i.e. Class)
y=df.iloc[:,4].values
y
```

```python
#Data Preprocessing
#Handle missing values
df.isnull().sum() #finding if missing values exist or not
```

```python
#Splitting dataset into train-test-split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_train
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
#Apply KNN
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
```

```python
model.predict([[5,2,3,4]])
```

```python
#Do testing of model with testing dataset
y_pred=model.predict(X_test)
```

```python
y_pred #predicted values
```

```python
y_test #original(test) values
```

```python
#Matching predicted & test values
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
cm 
```
